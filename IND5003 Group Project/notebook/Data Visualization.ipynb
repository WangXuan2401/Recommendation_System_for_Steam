```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
game_data = pd.read_csv('../data/games.csv')
game_data.head()
```

```python
recommendations_data = pd.read_csv('../data/recommendations.csv')
recommendations_data.head()
```

```python
users_data = pd.read_csv('../data/users.csv')
users_data.head()
```

```python
steam_data = pd.read_csv('../data/steam.csv')
steam_data.head()
```

```python
custom_order = [
    'Positive', 'Very Positive', 'Mixed', 'Mostly Positive', 'Mostly Negative',
    'Overwhelmingly Positive', 'Negative', 'Very Negative',
    'Overwhelmingly Negative'
]
ratings_count = game_data.rating.value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(y=custom_order, x=ratings_count, palette='crest_r', edgecolor='k')
plt.ylabel('rating', fontsize=10)
plt.xlabel('Games', fontsize=10)
plt.title('Steam review distribution')
plt.grid(True)
plt.show()
```
