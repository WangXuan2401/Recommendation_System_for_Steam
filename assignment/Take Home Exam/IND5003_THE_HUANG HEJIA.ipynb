**Use the data in the file `MRCI_clus_2310.csv`.**

Cluster the colleges on the five variables `ADM_RATE`, `SAT_AVG`, `C150_4`, `UGDS_WOMEN`, and `STUFACR` **after standardisation** using an **appropriately chosen** linkage method for **each of**

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read dataset
clus_data = pd.read_csv('MRCI_clus_2310.csv')
clus_data.head()
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standardized_data = scaler.fit_transform(clus_data)
clus_standardized = pd.DataFrame(
    standardized_data,
    columns=['ADM_RATE', 'SAT_AVG', 'C150_4', 'UGDS_WOMEN', 'STUFACR'])
```

```python
# Choose proper linkage method 
from scipy.cluster.hierarchy import linkage, dendrogram, cophenet
from scipy.spatial.distance import pdist, squareform

# Euclidean distance
euclidean_results = {}
euclidean_c_values = {}
euclidean_methods = ['single', 'complete', 'average', 'ward']
for method in euclidean_methods:
    Z = linkage(clus_standardized, method=method, metric='euclidean')
    c, _ = cophenet(Z, pdist(clus_standardized))
    euclidean_c_values[method] = c
    euclidean_results[method] = (Z, c)

# Manhattan distance
manhattan_results = {}
manhattan_c_values = {}
manhattan_methods = ['single', 'complete', 'average']
for method in manhattan_methods:
    Z = linkage(clus_standardized, method=method, metric='cityblock')
    c, _ = cophenet(Z, pdist(clus_standardized))
    manhattan_c_values[method] = c
    manhattan_results[method] = (Z, c)

# Visualize results    
manhattan_c_values['ward'] = None
cophenetic_data = {
    'Linkage Method': euclidean_c_values.keys(),
    'Euclidean Distance': euclidean_c_values.values(),
    'Manhattan Distance': manhattan_c_values.values()
}
cophenetic_df = pd.DataFrame(cophenetic_data)
cophenetic_df.set_index('Linkage Method', inplace=True)

plt.figure(figsize=(6, 4))
sns.heatmap(cophenetic_df, annot=True, cmap='YlGnBu', cbar=True, linewidths=0.5)
plt.title('Cophenetic Correlation Coefficients for Different Linkage Methods')
plt.show()
```

As can see in the plot, **Average Linkage** has the highest Cophenetic Correlation Coefficients for both **Euclidean distance** and 
**Manhattan distance**  
Cophenetic Correlation Coefficients for **Euclidean distance**: 0.77  
Cophenetic Correlation Coefficients for **Manhattan distance**: 0.71


The **cophenetic correlation coefficient** is a measure used to assess how faithfully a dendrogram preserves the pairwise distances between the original data points. The coefficient is a value between -1 and 1, where a value closer to 1 indicates that the dendrogram effectively maintains the original distances. It is calculated by comparing the actual pairwise distances of all samples to those implied by the hierarchical clustering.

After computing the hierarchical clustering for various linkage methods, we used the cophenetic correlation coefficient to determine which linkage method preserves the original distances best. The method that gives the highest cophenetic correlation is chosen as the most appropriate linkage method, as it ensures that the structure of clusters best reflects the original data's structure.

```python
# Plot the dendrograms for the linkage methods
fig, ax = plt.subplots(figsize=(15, 5))
Z, c = euclidean_results['average']
dendrogram(Z, ax=ax, truncate_mode='level', p=8)
ax.grid(False)
ax.set_title('Dendrogram of Average Linkage for Euclidean distance')
ax.set_xlabel("Clusters")
ax.set_ylabel("Distance")

fig, ax = plt.subplots(figsize=(15, 5))
Z, c = manhattan_results['average']
dendrogram(Z, ax=ax, truncate_mode='level', p=8)
ax.grid(False)
ax.set_title('Dendrogram of Average Linkage for Manhattan distance')
ax.set_xlabel("Clusters")
ax.set_ylabel("Distance")
plt.tight_layout()
plt.show()
```

```python
# Calculate silhouette scores
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster

# prepare for outline properties
euclidean_report = clus_standardized.copy()
manhattan_report = clus_standardized.copy()

# Euclidean distance
best_euclidean_score = -1
best_euclidean_clusters = 2
for clusters in range(2, 30):
    labels = fcluster(euclidean_results['average'][0], clusters, criterion='maxclust')
    score = silhouette_score(clus_standardized, labels, metric='euclidean')
    if score > best_euclidean_score:
        best_euclidean_score = score
        best_euclidean_clusters = clusters
        euclidean_report['Clusters'] = labels

# Manhattan distance
best_manhattan_score = -1
best_manhattan_clusters = 2
for clusters in range(2, 30):
    labels = fcluster(manhattan_results['average'][0], clusters, criterion='maxclust')
    score = silhouette_score(clus_standardized, labels, metric='cityblock')
    if score > best_manhattan_score:
        best_manhattan_score = score
        best_manhattan_clusters = clusters
        manhattan_report['Clusters'] = labels
        
# print the result
print('The optimal number of clusters for Euclidean distance is: ', best_euclidean_clusters)
print('The optimal number of clusters for Manhattan distance is: ', best_manhattan_clusters)
```

**Calculated silhouette scores and for both distance the optimal number of clusters is two.**

```python
# Helper function to outline the properties/differences of the clusters
def describe_clusters(Z, num_clusters, data, features, metric):
    # Get cluster labels
    labels = fcluster(Z, num_clusters, criterion='maxclust')
    
    # Add cluster labels to the dataset
    data['Cluster'] = labels
    
    # Display cluster properties
    cluster_description = data.groupby('Cluster')[features].mean()
    cluster_counts = data['Cluster'].value_counts().sort_index()
    
    print(f"Cluster Counts for {metric}:")
    print(cluster_counts)
    return cluster_description

# Describing clusters for both Euclidean and Manhattan distances
clus_features = ['ADM_RATE', 'SAT_AVG', 'C150_4', 'UGDS_WOMEN', 'STUFACR']
euclidean_clusters_desc = describe_clusters(euclidean_results['average'][0], best_euclidean_clusters, 
                                           clus_standardized.copy(), clus_features, 'Euclidean')
manhattan_clusters_desc = describe_clusters(manhattan_results['average'][0], best_manhattan_clusters, 
                                            clus_standardized.copy(), clus_features, 'Manhattan')
print('Description for euclidean clusters: ')
print(euclidean_clusters_desc) 
print('Description for manhattan clusters: ')
print(manhattan_clusters_desc)
```

**Result**  
For Euclidean Distance (Average Linkage):  
- Cluster 1 (855 colleges):
    Higher admission rate, lower SAT scores, lower completion rates, a slightly higher proportion of women, and a higher student-to-faculty ratio.  
- Cluster 2 (1 colleges):  
    Lower admission rate, higher SAT scores, higher completion rates, a slightly lower proportion of women, and a lower student-to-faculty ratio.  
- **Because using hierarchical clustering consider each point as its own cluster, the result of only one cluster in the group is normal.**  
- **As 2 clusters have the highest silhouette scores, we can more clusters for better interpretation.**
For Manhattan Distance (Average Linkage):  
- Cluster 1 (115 colleges):  
    Similar characteristics to the Euclidean Cluster 1, with lower admission rate, higher SAT scores, higher completion rates, a slightly lower proportion of women, and a lower student-to-faculty ratio.  
- Cluster 2 (741 colleges):  
    Similar characteristics to the Euclidean Cluster 2, with higher admission rate, lower SAT scores, lower completion rates, a slightly higher proportion of women, and a higher student-to-faculty ratio.  
      
These clusters generally segregate the colleges into two main categories: more selective institutions with higher academic performance metrics and less selective institutions with lower performance metrics.

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.outliers_influence as oinf
from sklearn.linear_model import LinearRegression

regr_data = pd.read_csv("MRCI_regr_2310.csv")
regr_data.head()
```

```python
# 1.Fit linear Regression model
X = regr_data[['ADM_RATE', 'SAT_AVG', 'UGDS', 'COSTT4_A', 'C150_4', 'UGDS_WOMEN', 'STUFACR']]
y = regr_data['MD_EARN_WNE_1YR']

X1 = sm.add_constant(X)
regr_OLS = sm.OLS(y, X1)
regr_OLS_results = regr_OLS.fit()
regr_OLS_results.summary()
```

```python
# 2. Study residuals
regr_OLS_results_inf = regr_OLS_results.get_influence()
regr_OLS_star = regr_OLS_results_inf.get_resid_studentized_external()
# claculate RMSPE
RMSPE = np.sqrt((regr_OLS_results_inf.resid_press**2).mean())
RMSPE
```

**4. Check for multicollinearity**  
**Multicollinearity:**  
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other, it implies that they convey similar information about the variance within the dataset.  
**Impact:**  
-Multicollinearity can severely affect the qualitative interpretation of the model.  
-It becomes challenging to decipher the individual impact of predictors, as a change in one is often associated with changes in others.  
-The p-values for the affected variables may not be reliable, leading to incorrect conclusions about the significance of the predictors.

```python
# 4. Check for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["Features"] = regr_OLS_results.model.exog_names[1:]
vif_data["VIF"] = [variance_inflation_factor(regr_OLS_results.model.exog, i) for i in range(1, regr_OLS_results.model.exog.shape[1])]
for index, row in vif_data.iterrows():
    print(f"VIF of {row['Features']}: {row['VIF']:.2f}")
```

***'SAT_AVG'** has a VIF of 4.40 and **'C150_4'** has a VIF of 3.74 indicates some level of multicollinearity.  
The remaining variables have VIF values below the common threshold, indicating that multicollinearity may not be significant for those variables.* 

**Findings:**
The model explains about 22.9% of the variance in MD_EARN_WNE_1YR.   
Among the predictors, SAT_AVG, C150_4, and UGDS_WOMEN are statistically significant.  
There are clearly some points with high influential to the model.  
There are some multicollinearity involving SAT_AVG and C150_4.

### **Q2.2: "Have you tried adding STATE to your explanatory variables?"**  
*Adding STATE to the model can help for better explain the model, however it can increase conditional number significantly.*

*One Hot Encoding STATE to fit the model*

```python
# Adding STATE to explanatory variables
model = regr_data.copy()
features = ['ADM_RATE', 'SAT_AVG', 'UGDS', 'COSTT4_A', 'C150_4', 'UGDS_WOMEN', 'STUFACR']

# log-transform other variables
cols_to_log = [col for col in features if (model[col] > 0).all()]
for col in cols_to_log:
    model[col] = np.log(model[col])

# One hot encode using get_dummies
model = pd.get_dummies(model, columns=['STATE'], drop_first=True)
model.columns = [col.replace(" ", "_") for col in model.columns]
# Fit model with STATE
formula_with_state = "MD_EARN_WNE_1YR ~ ADM_RATE + SAT_AVG + UGDS + COSTT4_A + C150_4 + UGDS_WOMEN + STUFACR + " + " + ".join(model.columns[model.columns.str.startswith("STATE_")])
model_with_state = smf.ols(formula_with_state, data=model).fit()

# Check for influential points using leverage and Cook's distance
model_with_state_cleaned_inf = model_with_state.get_influence()
(c, p) = model_with_state_cleaned_inf.cooks_distance
ids = range(len(model_with_state_cleaned_inf.cooks_distance[0]))
plt.scatter(ids, model_with_state_cleaned_inf.cooks_distance[0])
plt.title("Cook's Distance Plot")
plt.xlabel('id')
plt.ylabel("Cook's Distance")
plt.show()
```

```python
# Remove the most influential point
most_influential_point_index = np.argmax(c)
regr_data_cleaned = model.drop(most_influential_point_index)

# Re-fit the regression model without the influential point
model_with_state_cleaned = smf.ols(formula_with_state, data=regr_data_cleaned).fit()

# Display model summary for assessment
model_with_state_cleaned.summary()
```

```python
# Generate predictions and residuals
predictions_cleaned = model_with_state_cleaned.predict(regr_data_cleaned)
residuals_cleaned = regr_data_cleaned["MD_EARN_WNE_1YR"] - predictions_cleaned

# Set up the figure and axes for visualizations
fig, ax = plt.subplots(2, 2, figsize=(15, 12))

# Plot Residuals vs Fitted values for model_with_state_cleaned
sns.residplot(x=predictions_cleaned, y=residuals_cleaned, lowess=True, ax=ax[0, 0], line_kws={'color': 'red', 'lw': 1})
ax[0, 0].set_title('Residuals vs Fitted')
ax[0, 0].set_xlabel('Fitted values')
ax[0, 0].set_ylabel('Residuals')

# Histogram of Residuals for model_with_state_cleaned
sns.histplot(residuals_cleaned, kde=True, ax=ax[0, 1])
ax[0, 1].set_title('Histogram of Residuals')
ax[0, 1].set_xlabel('Residuals')

# Scatter plot of Actual vs Predicted values for model_with_state_cleaned
ax[1, 0].scatter(regr_data_cleaned["MD_EARN_WNE_1YR"], predictions_cleaned)
ax[1, 0].plot([regr_data_cleaned["MD_EARN_WNE_1YR"].min(), regr_data_cleaned["MD_EARN_WNE_1YR"].max()], 
             [regr_data_cleaned["MD_EARN_WNE_1YR"].min(), regr_data_cleaned["MD_EARN_WNE_1YR"].max()], 'r--', lw=2)
ax[1, 0].set_title('Actual vs Predicted (Cleaned Model with State)')
ax[1, 0].set_xlabel('Actual')
ax[1, 0].set_ylabel('Predicted')

# Coefficient Plot for model_with_state_cleaned
coefs_cleaned = pd.DataFrame({
    'Variable': model_with_state_cleaned.params.index[1:],  # Exclude intercept
    'Coefficient': model_with_state_cleaned.params.values[1:]
})
coefs_cleaned = coefs_cleaned.sort_values(by="Coefficient")
coefs_cleaned.plot(kind='barh', y='Coefficient', x='Variable', ax=ax[1, 1], color='blue', legend=False)
tick_positions = range(0, len(coefs_cleaned), 2)
ax[1, 1].set_yticks(tick_positions)
ax[1, 1].set_yticklabels(coefs_cleaned['Variable'][tick_positions])
ax[1, 1].set_title('Coefficient Plot')

plt.tight_layout()
plt.show()
```
